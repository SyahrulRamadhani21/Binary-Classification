```python
import numpy as np
import pandas as pd
from collections import Counter

# Membaca data latih/uji
train_data = pd.read_excel('https://github.com/SyahrulRamadhani21/Binary-Classification/blob/main/traintest.xlsx?raw=true', sheet_name='train')
test_data = pd.read_excel('https://github.com/SyahrulRamadhani21/Binary-Classification/blob/main/traintest.xlsx?raw=true', sheet_name='test')
```

```python
# Pelatihan atau training model
def euclidean_distance(vec1, vec2):
    return np.sqrt(np.sum(np.square(vec1 - vec2)))


# test_x = test_data.values[:, 1:-1][6]
all_distances = []
for test_x in test_data.values[:, 1:-1]:
    distance_list = []
    for train_x in train_data.values[:, 1:-1]:
        # for test_x in test_data.values[:, 1:-1]:
        distance_list.append(euclidean_distance(train_x, test_x))
    all_distances.append({'test_data': test_x, 'distance': distance_list})
# all_distances
```

```python
# Menyimpan model hasil training
all_model = []
for model in all_distances:
    # print(model)
    distance_df = pd.DataFrame(
        {"Distance": model['distance']}, index=train_data.index)
    result = pd.concat([train_data, distance_df],
                       axis=1).sort_values(by="Distance")
    all_model.append({'test_data': model['test_data'], 'test_model': result})
#all_model[0]['test_model']  # contoh salah satu modelnya di index ke 0
```

```python
# Evaluasi output
all_output = []
for model in all_model:
    output = Counter(model['test_model'].head(4)['y'])
    all_output.append({'test_data': model['test_data'], 'test_model': result,'y': max(output, key=output.get)})
```

```python
# Evaluasi model

for model in all_model:
    print(model['test_model'].head(4))
    print(output)
    print('Hasil:', model['test_data'], 'y =', max(output, key=output.get), '\n')
```
